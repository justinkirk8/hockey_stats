```python
# building loops to get info on all teams
x=0
y=0

season_list=['19171918','19181919','19191920','19201921','19211922','19221923','19231924','19241925','19251926','19261927','19271928','19281929','19291930','19301931','19311932','19321933','19331934','19341935','19351936','19361937','19371938','19381939','19391940','19401941','19411942','19421943','19431944','19441945','19451946','19461947','19471948','19481949','19491950','19501951','19511952','19521953','19531954','19541955','19551956','19561957','19571958','19581959','19591960','19601961','19611962','19621963','19631964','19641965','19651966','19661967','19671968','19681969','19691970','19701971','19711972','19721973','19731974','19741975','19751976','19761977','19771978','19781979','19791980','19801981','19811982','19821983','19831984','19841985','19851986','19861987','19871988','19881989','19891990','19901991','19911992','19921993','19931994','19941995','19951996','19961997','19971998','19981999','19992000','20002001','20012002','20022003','20032004','20042005','20052006','20062007','20072008','20082009','20092010','20102011','20112012','20122013','20132014','20142015','20152016','20162017','20172018','20182019','20192020','20202021']

historical_stat_df=pd.DataFrame(columns=('Team','Season','Conference','Division','GP','W','L','QTL','Pts','GF','GA','Diff'))


for season in season_list:
    
    query_url= f"https://statsapi.web.nhl.com/api/v1/standings?season={season}"
    hockey_response = requests.get(query_url)
    hockey_json = hockey_response.json()
    x = 0
    x_max = len(hockey_json['records'])

    while x < x_max:
        y = 0  
        y_max = len(hockey_json['records'][x]['teamRecords'])
        while y < y_max:
            try:
                team_name = hockey_json['records'][x]['teamRecords'][y]['team']['name']
            except:
                team_name = None
            
            try:
                team_division = hockey_json['records'][x]['division']['name']
            except:
                team_division = None
            
            try:
                team_conference = hockey_json['records'][x]['conference']['name']
            except:
                team_conference = None
            
            try:
                team_gp = hockey_json['records'][x]['teamRecords'][y]['gamesPlayed']
            except:
                team_gp = None
            
            try:
                team_W = hockey_json['records'][x]['teamRecords'][y]['leagueRecord']['wins']
            except:
                team_W = None
            
            try:
                team_L = hockey_json['records'][x]['teamRecords'][y]['leagueRecord']['losses']
            except:
                team_L = None
            
            try:
                team_QTL = hockey_json['records'][x]['teamRecords'][y]['leagueRecord']['ot']
            except:
                team_QTL = None
            
            try:
                team_Pts = hockey_json['records'][x]['teamRecords'][y]['points']
            except:
                team_Pts = None
            
            try:
                team_GF = hockey_json['records'][x]['teamRecords'][y]['goalsScored']
            except:
                team_GF = None
            
            try:
                team_GA = hockey_json['records'][x]['teamRecords'][y]['goalsAgainst']
            except:
                team_GA = None
            
            try:
                team_diff = team_GF - team_GA
            except:
                team_diff = None

            team = {'Team' : team_name,
               'Season' : season,
               'Division':team_division,
               'Conference':team_conference,
               'GP': team_gp,
               'W': team_W,
               'L': team_L,
               'QTL': team_QTL,
               'Pts': team_Pts,
               'GF': team_GF,
               'GA': team_GA,
               'Diff': team_diff}

            historical_stat_df = historical_stat_df.append(team,ignore_index=True)
            y = y + 1
        x = x + 1

historical_stat_df
```
